```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("magic04.data")
df
```

```python
cols = ["fLength", "fWidth", "fSize", "fConc", "fConc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
```

```python
df = pd.read_csv("magic04.data", names=cols)
df
```

```python
for i in df.index:
    if df.loc[i, 'class'] == 'g':
        df.loc[i, 'class'] = 1
    else:
        df.loc[i, 'class'] = 0
df
```

```python
plt.hist(df['fLength'], color = 'blue', label="fLength", alpha=0.7, density=True)
plt.title("fLength")
plt.legend()
plt.show()
```

```python
for label in df.columns:
    plt.hist(df[label], color = 'blue', label=label, alpha=0.7, density=True)
    plt.title(label)
    plt.legend()
    plt.show()
```

```python
plt.hist(df['fLength'], color = 'blue', label="fLength", alpha=0.7, density=True)
plt.hist(df['fWidth'], color = 'red', label="fWidth", alpha=0.7, density=True)
plt.title("fLength/fWidth")
plt.legend()
plt.show()
```
